```python
import pandas as pd
```

```python
df=pd.read_csv('/content/test.csv')
df
```

```python
y = df["three_g"]
x1 = df[["touch_screen","wifi"]]
```

```python
import statsmodels.api as sm
```

```python
x=sm.add_constant(x1)
x
```

```python
model = sm.Logit(y,x)
result = model.fit()
result.summary()
```

```python
df["three_g"].value_counts()
```

```python
cm_df =pd.DataFrame(result.pred_table())
cm_df.columns =["predicted 0", "predicted 1"]
cm_df.rename(index = {0:"Actual 0",1:"Actual 1"},inplace = True)
cm_df
```

```python
accuracy = (cm_df.iloc[0,0]+cm_df.iloc[1,1])/cm_df.sum().sum()
accuracy
```

```python
precision_0 =  cm_df.iloc[0,0] /(cm_df.iloc[0,0]+cm_df.iloc[1,0])
precision_0
```

```python
precision_1 =  cm_df.iloc[1,1] /(cm_df.iloc[1,1]+cm_df.iloc[0,1])
precision_1
```

```python
recall_0 = cm_df.iloc[0,0] / (cm_df.iloc[0,0] +cm_df.iloc[0,1])
recall_0
```

```python
recall_1 = cm_df.iloc[1,1] / (cm_df.iloc[1,1] +cm_df.iloc[1,0])
recall_1
```
